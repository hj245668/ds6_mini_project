[ feature engineering 1 ]

구매 금액
- z-score = (x - 평균) / 표준편차
- 평소보다 얼마나 많이 썼는지 (지나치게 많이 쓴 내용 확인할때 사용)

```python
# z-score 예시 데이터프레임
temp = pd.DataFrame({
    'a': [10, 20, 30, 20, 10, 200],
    'b': [100, 300, 200, 150, 250, 200],
    'c': [10, 500, 20, 250, 25, 200]
})

print(temp)
```

```python
temp.mean()
```

```python
temp.std()
```

```python
# 각각의 데이터에 대해 a 컬럼의 z-score를 계산합니다.
# (data - mean) / std

(temp['a'] - 48.33) / 74.67
```

```python
# Q. 각각의 데이터에 대해 b 컬럼의 z-score를 계산합니다.

(temp['b'] - 200.00) / 70.71
```

```python
# Q. 각각의 데이터에 대해 c 컬럼의 z-score를 계산합니다.

(temp['c'] - 167.50) / 192.50
```

```python
mean_vals = temp.mean()
std_vals = temp.std()

# z-score = (x - 평균) / 표준편차
z_score = (temp - mean_vals) / std_vals

result = pd.concat([temp, z_score.add_suffix('_z')], axis=1)
print(result.round(3))
```

```python
# <---------------------------------------
# <--- a 가 200 만원이 지나치게 많이 썼음
```

시간 분석

```python
# Q. trans_date_trans_time 컬럼에서 시간 값을 추출하여 hour 컬럼에 저장합니다.
# (힌트: datetime 자료형을 이용합니다.)
```

```python
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()
```

```python
cc_df['hour'].describe()
```

```python
# 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수를 정의합니다.
# 입맛에 맞춰서 임의로 조정 가능

def hour_func(x):
    x = x % 24                # 24시 -> morning
    if 0 <= x < 12:
        return 'morning'      # dawn + morning
    elif 12 <= x < 18:
        return 'afternoon'
    else:
        return 'night'
```

```python
# Q. hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장합니다.

cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
# Q. 시간대별로 데이터 수를 계산하여 출력합니다.

cc_df['hour_cat'].value_counts()
```

```python
# <----------------------------------
# 카드별 총 몇건이고 아침점심저녁 각각이 결제건이 얼마나 되는지 확인해봅니다.
```

```python
# cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장합니다.

all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
# cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장합니다.

hour_cnt = cc_df.groupby(['cc_num','hour_cat'])['amt'].count().reset_index()
```

```python
# Q. cc_num 컬럼을 기준으로 hour_cnt와 all_cnt를 합칩니다.
# (hour_cnt에서 left merge를 수행합니다.

hour_cnt = hour_cnt.merge(all_cnt, on = 'cc_num', how = 'left')
```

```python
# Q. amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경합니다.

hour_cnt.rename({'amt_x' : 'hour_cnt', 'amt_y'  : 'total_cnt'},  axis = 1)
```

```python
"""
====                  cc_df['distance'] in Aiffel                   ====
==== Aiffel 실행 안 (커널 동작 오류로 distance 계산법을 변경사용함) ====


참고 site : https://stackoverflow.com/questions/4913349/haversine-formula-in-python-bearing-and-distance-between-two-gps-points?utm_source=chatgpt.com

# 고객의 위치와 상점의 위치 사이의 거리를 계산하여 distance 컬럼에 저장합니다.
# (모든 데이터가 계산되므로 실행 시간이 오래 걸릴 수 있습니다.)
# cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)

# (함수호출 시간 오래걸려서 벡터화해서 실행합니다.Haversine 공식 적용)
# 지구를 위도(latitude)·경도(longitude)를 구면좌표로 보고
# 두 점의 최단거리(대원거리, great-circle distance)를 구하는 공식
import numpy as np

# 지구 반지름
R = 6371

# 위도·경도를 라디안 단위로 변환
lat1 = np.radians(cc_df['lat'].to_numpy())
lon1 = np.radians(cc_df['long'].to_numpy())
lat2 = np.radians(cc_df['merch_lat'].to_numpy())
lon2 = np.radians(cc_df['merch_long'].to_numpy())

# 위도·경도 차이 계산 (lat2 - lat1)/2 , (lon2 - lon1)/2
# Haversine 식 계산 (구면 삼각법)
# 최종 거리 공식 : R * 2 * np.arcsin(np.sqrt(...))
cc_df['distance'] = R * 2 * np.arcsin(np.sqrt(
    np.sin((lat2 - lat1)/2)**2 +
    np.cos(lat1) * np.cos(lat2) * np.sin((lon2 - lon1)/2)**2
))
"""
```
