```python
# 추려진 변수들로 상관관계 시각화 (히트맵)

import matplotlib.pyplot as plt
import seaborn as sns

# 상관관계 높은 10개 데이터만 추려서 컬럼 추출
corr_cols = bonus_df_encoded.corr()['Price'].abs().sort_values(ascending=False).head(10).index
corr_subset = bonus_df_encoded[corr_cols].corr()

plt.figure(figsize=(8, 8))
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Top 10 correlated features with Price')
plt.show()
```

```python
# 상관성이 높은 5개 변수를 추려 시각화

# 1. 상관관계 상위 5개
top_corr_cols = bonus_df_encoded.corr()['Price'].abs().sort_values(ascending=False).head(6).index

# 2. 수치형태만 필터링
pair_df = bonus_df_encoded[top_corr_cols].select_dtypes(include=['number'])

# 3. 데이터 많을때 일부 샘플링
pair_df = pair_df.sample(400, random_state=42)

# 4. Pairplot 시각화
sns.pairplot(pair_df, diag_kind='kde', corner=True)
plt.suptitle('Pairwise Relationships among Top 5 Correlated Features with Price', y=1.02)
plt.show()
```

```python
# Emission Class_Euro 6 가 안나옴. 원인 : one hot encoding 으로 data type bool형
# 수치화 하여 시각화 ->
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 상관관계 상위 5개
top_corr_cols = bonus_df_encoded.corr()['Price'].abs().sort_values(ascending=False).head(6).index

# 2. 선택한 컬럼만 추출
pair_df = bonus_df_encoded[top_corr_cols].copy()

# 3. bool 컬럼을 int로 변환 (True→1, False→0)
for col in pair_df.columns:
    if pair_df[col].dtype == 'bool':
        pair_df[col] = pair_df[col].astype(int)

# 4.샘플링 후 시각화
pair_df = pair_df.sample(400, random_state=42)

sns.pairplot(pair_df, diag_kind='kde', corner=True)
plt.suptitle('Pairwise Relationships among Top 5 Correlated Features with Price', y=1.02)
plt.show()
```

##### 1. Pairplot
#####   1_1. 대각선: 데이터 분포(Distribution)

- Registration_Year : 오른쪽 치우친 분포 -> 최근 연식 차량이 많음

- Mileage(miles) : 왼쪽으로 치우친 분포 -> 20만 km 이하

- Previous Owners : 1~3명 구간 집중 -> 개인 소유  차량 이력

- Emission Class_Euro 6 vs 4 : 두 피크가 명확히 구분됨 -> 최신 vs 구형 등급 차량이 명확하게 구분됨


#####   1_2. 대각선 이하 : 산점도(Scatter)
- 오른쪽 위로 퍼져 있으면 양의 상관
- 왼쪽 위→오른쪽 아래로 퍼져 있으면 음의 상관

#####     1_2_2. 상관관계
- Price ↔ Registration_Year (강한 양의 상관) 최근 등록 차량일 수록 가격이 높음
- Price ↔ Mileage(miles) (강한 음의 상관) 주행거리(Mileage) 많을수록 차량 가격이 낮음
- Price ↔ Previous Owners (약한 음의 상관) 소유자 수 많을수록 가격이 낮아짐
- Price ↔ Emission Class_Euro 6 (양의 상관) 최신 배출 기준(Euro 6) 만족하는 차량일수록 가격이 높음
- Price ↔ Emission Class_Euro 4 (음의 상관) 낮은 등급(Euro 4) 차량일수록 구형차로, 가격 하락


##### 2. Price를 기준으로,
- 가격 상승 요인 : 최근 연식, 최신 배출 등급
- 가격 하락 요인 : 주행 거리 많음. 소유자 많음. 배출 등급 낮음
